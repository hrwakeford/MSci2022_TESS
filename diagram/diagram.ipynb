```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('exoplanets.csv')

period = df["PER"]
radius = df["R"]

xlog = np.log(period)
ylog = np.log(radius)
```

```python
fig, ax = plt.subplots()
ax.set_xscale('log', base=10) 
ax.set_yscale('log', base=10) 

ax1 = ax.twinx()
ax3 = ax.twiny()


ax3.set_xscale('log', base=10) 
ax1.set_ylim(ax.get_ylim())
ax3.set_xlim(ax.get_xlim())


ax.scatter(df["PER"], df["R"], s=5, color="black")
ax.scatter(df.loc[df["EANAME"]=="WASP-17 b"]["PER"], df.loc[df["EANAME"]=="WASP-17 b"]["R"], label="WASP-17b") 
ax.scatter(df.loc[df["EANAME"]=="WASP-39 b"]["PER"], df.loc[df["EANAME"]=="WASP-39 b"]["R"], label="WASP-39b", marker="s") 
ax.scatter(df.loc[df["EANAME"]=="HAT-P-26 b"]["PER"], df.loc[df["EANAME"]=="HAT-P-26 b"]["R"], label="HAT-P-26b", marker="^")



ax.axhline(1, linestyle='--', color="gray") #Jupiter radius
ax.axhline(0.267644, linestyle='--', color="gray") #Sub-neptune
ax.axhline(0.0892147, linestyle='--', color="gray") #Earth-sized
ax.annotate("Jupiter-sized", xy=(0.07, 0.76), color="gray", weight="semibold")
ax.annotate("Sub-neptune", xy=(0.07, 0.205), color="gray", weight="semibold")
ax.annotate("Earth-sized", xy=(0.07, 0.066), color="gray", weight="semibold")


ax.set_xlabel("Orbital period (days)")
ax.set_ylabel("Planet radius (Jupiter radii)")
 
ax1.set_xscale('log',base=10) 
ax1.set_yscale('log',base=10) 

ax.set_xlim(0.065, 1000)
ax.set_ylim(0.012, 100)


ax1.set_ylim(ax.get_ylim())
ax3.set_xlim(ax.get_xlim())


ax1.set_yticklabels([])
ax3.set_xticklabels([])

ax.tick_params(which='both', direction = 'in', width=1.3, length=7)
ax1.tick_params(which='both', direction = 'in', width=1.3, length=7)
ax3.tick_params(which='both', direction = 'in', width=1.3, length=7)

ax.tick_params(which='minor', length=2.8)
ax1.tick_params(which='minor', length=2.8)
ax3.tick_params(which='minor', length=2.8)

ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda y,pos: ('{{:.{:1d}f}}'.format(int(np.maximum(-np.log10(y),0)))).format(y)))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y,pos: ('{{:.{:1d}f}}'.format(int(np.maximum(-np.log10(y),0)))).format(y)))

ax.legend()

plt.show()
```
