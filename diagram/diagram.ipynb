```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('exoplanets.csv')

period = df["PER"]
radius = df["R"]

xlog = np.log(period)
ylog = np.log(radius)
```

```python

plt.scatter(xlog, ylog, s=5, color="black")

plt.axhline(0.0274, linestyle='--', color="gray", label="Sub-neptune")
plt.axhline(0.0912, linestyle='--', color="gray", label="Earth-sized")

plt.xlabel("Orbital period (days)")
plt.ylabel("Planet radius (Jupiter radii)")
plt.xscale('log',base=10) 
plt.yscale('log',base=10) 
plt.show()
```
